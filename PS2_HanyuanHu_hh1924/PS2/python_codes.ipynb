```python
import pandas as pd
import numpy as np
```

```python
from scipy.stats import expon
```

```python
TOTAL_SIM = 1_000_000
```

```python
# CPU
df = pd.read_csv("CPU_exp.csv").iloc[:, 1]
axs = df.hist(bins=1000, cumulative=True)

x = np.linspace(0, 16, 100000)
y = expon.cdf(x) * TOTAL_SIM
axs.plot(x, y, "r", linewidth=4);
```
